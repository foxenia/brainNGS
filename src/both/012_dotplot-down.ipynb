```r
deseq_gdiff = readRDS('../../rds/deseq2_gut_diff_mockvirus.rds')
deseq_gundiff = readRDS('../../rds/deseq2_gut_undiff_mockvirus.rds')
deseq_brain = readRDS('../../rds/deseq2_brain_mockvirus.rds')
```

```r
# Combine the genes using union
gene_uni <- union(rownames(deseq_gundiff$MLB2), rownames(deseq_gdiff$MLB2))  # Combine first two datasets
gene_uni <- union(gene_uni, rownames(deseq_brain$MLB2))  # Add the third dataset

# View the result
head(gene_uni)
length(gene_uni)
```

```r
#checking the 
# Combine the genes using union
gene_uni_VA1 <- union(rownames(deseq_gundiff$VA1), rownames(deseq_gdiff$VA1))  # Combine first two datasets
gene_uni_VA1 <- union(gene_uni, rownames(deseq_brain$VA1))  # Add the third dataset

# View the result
head(gene_uni_VA1)
length(gene_uni_VA1)
```

```r
gundiff_sgn_genes = lapply(deseq_gundiff,function(x)rownames(x)[!is.na(x$padj) & x$padj<0.05 & x$log2FoldChange > 1])
```

```r
gdiff_sgn_genes = lapply(deseq_gdiff,function(x)rownames(x)[!is.na(x$padj) & x$padj<0.05 & x$log2FoldChange > 1])
```

```r
b_sgn_genes = lapply(deseq_brain,function(x)rownames(x)[!is.na(x$padj) & x$padj<0.2])
```

```r
gundiff_go = compareCluster(gundiff_sgn_genes,
                       fun='enrichGO',
                       universe      = gene_uni,
                       pAdjustMethod = "BH",
                       ont='ALL',
                       OrgDb = 'org.Hs.eg.db',
                       keyType = 'ENSEMBL')
```

```r
gdiff_go = compareCluster(gdiff_sgn_genes,
                       fun='enrichGO',
                       universe      = gene_uni,
                       pAdjustMethod = "BH",
                       ont='ALL',
                       OrgDb = 'org.Hs.eg.db',
                       keyType = 'ENSEMBL')
```

```r
b_go = compareCluster(b_sgn_genes,
                       fun='enrichGO',
                       universe      = gene_uni,
                       pAdjustMethod = "BH",
                       ont='ALL',
                       OrgDb = 'org.Hs.eg.db',
                       keyType = 'ENSEMBL')
```
