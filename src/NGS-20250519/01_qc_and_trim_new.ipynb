# Create sample table
R

```python
# 1. Найдём все fq.gz файлы рекурсивно в подкаталогах VL*
files <- list.files(
  path = "../../data/NGS-20250519/01.RawData/",
  pattern = "\\.fq\\.gz$",
  recursive = TRUE,
  full.names = TRUE
)

# 2. Извлечём короткие имена: VL25_R1, VL25_R2 и т.д.
get_sample_alias <- function(path) {
  fname <- basename(path)
  sample_id <- sub("^(VL\\d+).*", "\\1", fname)
  read <- if (grepl("_1\\.fq\\.gz$", fname)) "R1" else if (grepl("_2\\.fq\\.gz$", fname)) "R2" else "R?"
  paste0(sample_id, "_", read)
}

# 3. Соберём таблицу без повторного склеивания путей
samples <- data.frame(
  sample_name = sapply(files, get_sample_alias),
  filepath = normalizePath(files),  # корректный абсолютный путь
  stringsAsFactors = FALSE
)

# 4. Проверим, что все пути существуют
print(any(!file.exists(samples$filepath)))  # должно быть FALSE
```

```python
write.csv(samples,"samples.csv")
```

# FastQC and MultiQC
(change kernel to python if use bash (it is not working in R one))

```python
#it's in R
dir.create("fastqc_results", showWarnings = FALSE)

apply(samples, 1, function(row) {
  cmd <- paste("fastqc -o fastqc_results", shQuote(row["filepath"]))
  message("Запускаем: ", cmd)
  system(cmd)
})
```

# Trimming - I will not trim this data!


Trimming - based on guidelines here - https://www.lexogen.com/wp-content/uploads/2019/12/015UG108V0140-QuantSeq-Data-Analysis-Pipeline_2017-07-03.pdf

polyA.fa.gz is https://github.com/BioInfoTools/BBMap/raw/master/resources/polyA.fa.gz, truseq.fa.gz is https://github.com/BioInfoTools/BBMap/raw/master/resources/truseq.fa.gz, both downloaded 16/01/25. truseq is the truseq adapters, polyA is just AAAAAAAA in a fasta file

    ref - reference file containing kmers to trim
    k=13 - Kmer length, looks for matches >= 13nt
    ktrim=r - Trim reads to remove bases matching reference kmers
    useshortkmers=t - Allow shorter kmers at read ends
    mink=5 - shorter kmers at read ends should be >=5nt
    qtrim=r - quality trim reads, right end only
    trimq=10 - minimum score to quality trim
    minlength=20 - discard reads <20nt after trimming
